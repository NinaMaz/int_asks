### Loading data

```python
import pandas as pd
```

```python
data = pd.read_csv("test.csv")
```

### How many unique parameters do users have

```python
data.groupby(['user_id'])['parameter_name'].unique().apply(lambda x: len(x))
```

### What is the average value of each parameters per user?

```python
data.groupby(['user_id', 'parameter_name'])['parameter_value'].mean()
```

### Do pivoting such that single user entry will have all his parameter in single row.

```python
data.pivot_table(index='user_id', columns='parameter_name', values='parameter_value')#.head()
```

### Resample data by week and find average values of each parameters per user

```python
data['measure_date'] = pd.to_datetime(data['measure_date'])
```

```python
data.groupby(['user_id', 'parameter_name']).resample('W', on='measure_date').mean()
```

**if we want a more concise looking table we can drop NaNs**

```python
data.groupby(['user_id', 'parameter_name']).resample('W', on='measure_date').mean().dropna()
```
